## VI. Sparse data handling

```python
# sparse coloumn creation by setting some of the values to be nan
wine.loc[np.random.choice(wine['magnesium'].count(), int(0.13*wine['magnesium'].count()), replace=False),'magnesium'] = np.nan
```

```python
# nan values check
np.where(wine['magnesium'].isna())
```

```python
# delete rows with any of the coloumns to be nan
wine.dropna()
```

```python
# or fill nan values with the mean value for the corresponding coloumn
wine.fillna({'magnesium':wine['magnesium'].mean()})
```
